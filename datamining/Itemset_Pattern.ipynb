Recordar que se cuenta con 50 clientes que pueden escoger entre un catalogo de 20 items. En el dataframe se muestra cada una de las transacciones (T1..T50) con los respectivos códigos de items adquiridos

```
import pandas as pd
from pymining import itemmining
from pymining import assocrules

df = pd.read_csv('https://drive.google.com/uc?id=15S4lUzXkT0J6xZksZJTCtmdFPJIVbTRS&export=download', header=None)
df
```

Vemos que los clientes T5,T49,T12 y T41 son los clientes que han comprado una mayor variedad de productos

```
#Se obtienen las transacciones con el mayor número de items comprados
serie = df.groupby(by=0).count().sum(axis=1).sort_values(ascending= False)
serie
```

El promedio de items comprados por cada cliente es menos de 5 (4.74)

```
#Promedio de numero de items por transaccion
promedio = serie.sum()/len(serie)
promedio
```

Se sabe que en cada transacción, las columnas de la 1 a la 8 indican los códigos de los items adquiridos. Así se tiene que los items 20, 11 y 1 son los más vendidos con 22, 18, 16 ocasiones, respectivamente

```
#El producto mas vendido. 
cols=[1,2,3,4,5,6,7,8]
suma = pd.Series()
for i in cols:
  res = df[i].value_counts()  
#Suma de series
  suma = suma.add(res, fill_value=0)
suma.sort_values(ascending=False)
```

En resumen, se obtiene que el más vendido es el **item 20** con 22 unidades y el  menos vendido es el **item 0** con 2 unidades

##Transformación de estructuras para el uso de ItemMining

```
#Se transforma el dataframe al formato de tupla/lista considerando las columnas que contiene los items comprados
cols = [1,2,3,4,5,6,7,8]
transacciones = tuple(df[cols].to_records(index=False).tolist())
#Es necesario eliminar los valores 'nan' para poder procesar posteriormente las reglas de asociacion
trans = []   #lista vacia

for i in range(0,len(transacciones)):
# Convierte a serie cada uno de los registros y elimina los 'nan'. Los añade a una lista de tuplas
  trans.append(tuple(pd.Series(transacciones[i]).dropna()))  
#Ahora convierte la lista de tuplas en una tupla de tuplas  
trans = tuple(trans)
trans
```

En el paso anterior tenemos la información en el formato necesario para aplicar datamining y encontrar *insights*. Para estos se van a identificar aquellos items o combinación de items que aparezcan con mayor frecuencia sobre la base de valor mínimo (min_support)

```
#Se extrae el conjunto de itemset frecuente
relim_input = itemmining.get_relim_input(trans)
#min_support es el "sigma" (soporte minimo absoluto) -> Los elementos del itemset deben aparecer al menos N veces en las transacciones
min_support = 5
item_sets = itemmining.relim(relim_input, min_support=min_support)

for itemset in item_sets:
  print (itemset)
```

```
#se construyen las reglas de asociacion con un nivel de confianza X
min_confidence = 0.7
rules = assocrules.mine_assoc_rules(item_sets, min_support=min_support, min_confidence=min_confidence)
for rule in rules:
  print (rule)
```

La regla de asociación permite identificar patrones de consumo entre los diferentes *itemsets*. Para esto se define un unmbral de soporte mínimo y un intervalo de confianza.
Por ejemplo, en  este caso, se tiene que el **item 7** origina el consumo del **item 20** con una certeza de **81.82%**.
De igual manera, se puede puede concluir que el consumo de los **items 8 y 20** da como como resultado, la compra del **item 1** con una certeza de **71.43%**. Estos patrones se dan en un soporte mínimo de frecuencia de 5

```
#Se convierte el objeto lista que contiene la regla de asociacion a DataFrame
df_rule=pd.DataFrame(rules)
#Se procede a ordenar por el nivel de confianza y luego por la cantidad de incidencias que se cumple la regla de asociacion
df_rule.sort_values(ascending=False,by=[3,2])
```
